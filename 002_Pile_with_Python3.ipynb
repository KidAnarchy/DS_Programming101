```python
# 002_Pile_with_Python3
import random    # Use random function for Player2(Computer)

print("\n======Take stick from pile Human VS Computer=======")
N = int(input('How many sticks (N) in the pile:'))   # N = amount of stick in the pile
print('There are', N, 'sticks in the pile.')
name = str(input('What is your name :'))             # Player 1 Somchai or Your name:
print('--------------------------------------------------')

for X in range (N):   # X = take a stick in the pile by Player1
    print('\n')
    print(name ,',how many sticks you will take (1 or 2):') 
    X = int(input())
    
# Constraint.
    if X >=3:
        print('No you cannot take more than 2 sticks!')
        
    elif 1<= X <3:
        
        if X == 2:
            N = N - X  
            if N==0:
                print ('\nYou loss,',name,'Are you crazy that you do like this?')
                break
            elif N==1:
                if N <= 1:
                    print ('\nI, smart computer, take the last stick.')
                    print (name, 'win (I, smart computer, am sad T_T)')
                    break
        elif X == 1:
            N = N - X  
            if N <= 0:
                print('\n',name,',take the last stick.')
                print(name, 'A smart computer, win!!!')
                break
            elif N == 1:
                print ('\nI, smart computer, take the last stick.')
                print (name, 'win (I, smart computer, am sad T_T)')
                break
        else:
            print('There are not enough sticks to take.')
            
        print('There are', N, 'sticks in the pile.')   
        
    else:
        print('No you cannot take less than 1 stick!')
        continue
    
# Player 2 Python the smartest computer:   
    for Y in range (N):             # Y = take a sticl in the pile by Computer
        Y = random.randint(1,2)     # This version we will use random function to pick stick from pile.
        print('\nI, smart computer take :',Y)
        N = N - Y
        if N > 0:
            break
        else:
            print('\nI,smart computer very sad that I do like this')
            print('Game Over')
            break
    print('There are', N, 'sticks in the pile.')
    print('----------------------------------------------------')
    
    if N == 0:
        break 
print('\n----------------------------------------------------')
print('----------------------------------------------------')
```
